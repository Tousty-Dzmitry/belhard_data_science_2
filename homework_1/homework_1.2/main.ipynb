```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import class_weight

# === Подготовка данных ===
X_train = df_train.drop('Class', axis=1)
y_train = df_train['Class']

X_test = df_test.drop('Class', axis=1)
y_test = df_test['Class']

# === Настройка весов классов ===
weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(zip(np.unique(y_train), weights))

# Усиливаем важность классов 1 и 2 (увеличиваем их вес)
class_weights[1] *= 1.5
class_weights[2] *= 1.5
# class_weights[1] *= 3.0
# class_weights[2] *= 3.0


sample_weights = np.array([class_weights[label] for label in y_train])

# === Настройка модели ===
model = XGBClassifier(
    objective='multi:softmax',
    num_class=3,
    eval_metric='mlogloss',
    n_estimators=500,
    learning_rate=0.05,
    # max_depth=5,
    max_depth=5,
    # min_child_weight=3,
    min_child_weight=3,
    # gamma=0.2,
    gamma=0.2,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    tree_method='hist',
    device='cpu',
    # scale_pos_weight=[1, 5, 5]  # Веса для классов 0, 1, 2
    # early_stopping_rounds убран, т.к. нет валидации
)




# === Обучение модели с весами ===
model.fit(
    X_train, y_train,
    sample_weight=sample_weights,
    verbose=True
)

# === Оценка на тестовом наборе ===
print("\n=== Test Metrics ===")
y_test_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, y_test_pred))

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_test_pred),
            annot=True, fmt='d', cmap='Greens',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.title('Test Confusion Matrix')
plt.show()
```

```python
# === Оценка на тестовом наборе ===
print("\n=== Test Metrics ===")
y_test_pred = model.predict(X_test)

### === Блок постобработки с акцентом на классы 0 и 1 === ###
y_proba = model.predict_proba(X_test)

# Усиливаем классы 0 и 1
class0_boost = 5.0  # Усиливаем класс 0
class1_boost = 2.0  # Усиливаем класс 1
class2_boost = 1.0  # Без изменений

# Применяем бустинг
y_proba[:, 0] *= class0_boost
y_proba[:, 1] *= class1_boost
y_proba[:, 2] *= class2_boost

# Нормализуем вероятности
y_proba = y_proba / y_proba.sum(axis=1)[:, np.newaxis]

# Новые предсказания
y_test_pred_boosted = np.argmax(y_proba, axis=1)
y_test_pred = y_test_pred_boosted
### === Конец блока === ###

# Выводим метрики
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")
print("Classification Report (Class 0 & 1 Focused):")
print(classification_report(y_test, y_test_pred, digits=4))

# Матрица ошибок
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_test_pred),
            annot=True, fmt='d', cmap='Blues',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title(f'Confusion Matrix (Boosts: Class 0 = {class0_boost}x, Class 1 = {class1_boost}x)')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# === Оценка на тестовом наборе ===
print("\n=== Test Metrics ===")
y_test_pred = model.predict(X_test)

### === Блок с фокусом на классы 1 и 2 === ###
y_proba = model.predict_proba(X_test)

# Коэффициенты усиления:
boost_factors = {
    0: 1.0,   # Класс 0 без изменений
    1: 2.5,   # Усиливаем класс 1
    2: 3.5    # Сильнее усиливаем класс 2
}

# Применяем усиление по классам
for class_idx, boost in boost_factors.items():
    y_proba[:, class_idx] *= boost

# Нормализация вероятностей (по строкам)
y_proba = y_proba / y_proba.sum(axis=1, keepdims=True)

# Новые предсказания
y_test_pred_boosted = np.argmax(y_proba, axis=1)

### === Метрики и визуализация === ###
print("\nClassification Report (Class 1 & 2 Boosted):")
print(classification_report(y_test, y_test_pred_boosted, digits=4))

# Матрица ошибок
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_test_pred_boosted),
            annot=True, fmt='d', cmap='Reds',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.title(f'Confusion Matrix (Class 1 Boost={boost_factors[1]}x, Class 2 Boost={boost_factors[2]}x)')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# ансамбль: три модели + базовая модель
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import class_weight


# === Ансамбль с базовой моделью и усилениями классов ===
base_proba = model.predict_proba(X_test)

# Усиление класса 0
proba_0 = base_proba.copy()
proba_0[:, 0] *= 2.0
proba_0 = proba_0 / proba_0.sum(axis=1)[:, np.newaxis]

# Усиление класса 1 и немного класса 2
proba_1 = base_proba.copy()
proba_1[:, 1] *= 3.0
proba_1[:, 2] *= 1.2
proba_1 = proba_1 / proba_1.sum(axis=1)[:, np.newaxis]

# Усиление класса 2
proba_2 = base_proba.copy()
proba_2[:, 2] *= 3.5
proba_2 = proba_2 / proba_2.sum(axis=1)[:, np.newaxis]

# === Веса моделей в ансамбле ===
w_base, w0, w1, w2 = 1.5, 2.2, 2.1, 1.7

# Взвешенное усреднение вероятностей (базовая + 3 усиленные версии)
ensemble_proba = (w_base * base_proba + w0 * proba_0 + w1 * proba_1 + w2 * proba_2) / (w_base + w0 + w1 + w2)

# Финальные предсказания ансамбля
ensemble_preds = np.argmax(ensemble_proba, axis=1)

# Метрики ансамбля
print("\n=== Ensemble Test Metrics (Base + Weighted Ensemble) ===")
print(f"Accuracy: {accuracy_score(y_test, ensemble_preds):.4f}")
print("Classification Report (Weighted Ensemble):")
print(classification_report(y_test, ensemble_preds, digits=4))

plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, ensemble_preds),
            annot=True, fmt='d', cmap='Purples',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.title('Confusion Matrix (Ensemble with Base Model)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

```
