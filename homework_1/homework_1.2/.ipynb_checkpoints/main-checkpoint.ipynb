```python
import matplotlib.pyplot as plt
import seaborn as sns

# Предполагая, что df_train - ваш DataFrame и 'Class' - столбец с метками классов
class_counts = df_train['Class'].value_counts()

# Создаем фигуру
plt.figure(figsize=(12, 6))

# Строим barplot с Seaborn для красивого отображения
sns.barplot(x=class_counts.index, 
            y=class_counts.values, 
            palette="viridis")

# Добавляем аннотации с количеством
for i, count in enumerate(class_counts.values):
    plt.text(i, count + 0.02*max(class_counts), 
             str(count), 
             ha='center', 
             va='bottom')

# Настройки графика
plt.title('Распределение классов в наборе данных', fontsize=16)
plt.xlabel('Класс', fontsize=14)
plt.ylabel('Количество образцов', fontsize=14)
plt.xticks(rotation=45)  # Поворачиваем подписи если нужно
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Увеличиваем отступ снизу если подписи длинные
plt.tight_layout()

# Показываем график
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE


# Отделяем признаки (X) и целевую переменную (y)
X = df_train.drop('Class', axis=1)  # Все столбцы, кроме 'Class'
y = df_train['Class']              # Только 'Class'

# Применяем SMOTE только к тренировочным данным
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Разделяем на train (80%) и validation (20%)
X_train, X_val, y_train, y_val = train_test_split(
    X, 
    y, 
    test_size=0.2,       # 20% в валидацию
    random_state=42,     # Для воспроизводимости
    stratify=y           # Сохраняет баланс классов (если есть дисбаланс)
)

# Тестовый набор (если df_test уже есть)
X_test = df_test.drop('Class', axis=1)  # Если 'Class' есть в df_test
y_test = df_test['Class']               # Если нужны истинные значения для оценки
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Создание и обучение CatBoost с валидационным набором
clf = CatBoostClassifier(
    iterations=500,               # Увеличил для более устойчивого обучения
    learning_rate=0.05,           # Меньше LR для более плавного обучения
    depth=6,                      # Немного глубже деревья
    random_state=42,
    verbose=10,                   # Показывать прогресс каждые 10 итераций
    eval_metric='Accuracy',       # Метрика для валидации
    early_stopping_rounds=20,     # Ранняя остановка если нет улучшений
    task_type='CPU',              # Можно изменить на GPU если доступно
    cat_features=None             # Укажите список категориальных фичей если есть
)

# Обучение с мониторингом на валидационном наборе
clf.fit(
    X_train, y_train,
    eval_set=(X_val, y_val),
    plot=True                    # Визуализация процесса обучения
)

# 1. Оценка на валидационном наборе
print("\n=== Validation Metrics ===")
y_val_pred = clf.predict(X_val)
print(f"Accuracy: {accuracy_score(y_val, y_val_pred):.4f}")
print("Classification Report:")
print(classification_report(y_val, y_val_pred))

# Матрица ошибок для валидации
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_val, y_val_pred),
            annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Validation Confusion Matrix')
plt.show()

# 2. Финальная оценка на тестовом наборе
print("\n=== Test Metrics ===")
y_test_pred = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, y_test_pred))

# Матрица ошибок для теста
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_test_pred),
            annot=True, fmt='d', cmap='Greens',
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Test Confusion Matrix')
plt.show()

# 3. Важные фичи (если нужно)
if len(X_train.columns) <= 30:  # Показываем только если фичей не слишком много
    plt.figure(figsize=(10, 6))
    sns.barplot(x=clf.feature_importances_, y=X_train.columns)
    plt.title('Feature Importance')
    plt.show()
```
