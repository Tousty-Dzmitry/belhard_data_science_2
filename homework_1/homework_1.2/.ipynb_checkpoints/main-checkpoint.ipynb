```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE


#   (X)    (y)
X = df_train.drop('Class', axis=1)  #  ,  'Class'
y = df_train['Class']              #  'Class'

#  SMOTE    
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

#   train (80%)  validation (20%)
X_train, X_val, y_train, y_val = train_test_split(
    X, 
    y, 
    test_size=0.2,       # 20%  
    random_state=42,     #  
    stratify=y           #    (  )
)

#   ( df_test  )
X_test = df_test.drop('Class', axis=1)  #  'Class'   df_test
y_test = df_test['Class']               #      
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


#    CatBoost   
clf = CatBoostClassifier(
    iterations=500,               #     
    learning_rate=0.05,           #  LR    
    depth=6,                      #   
    random_state=42,
    verbose=10,                   #    10 
    eval_metric='Accuracy',       #   
    early_stopping_rounds=20,     #     
    task_type='CPU',              #    GPU  
    cat_features=None             #      
)

#      
clf.fit(
    X_train, y_train,
    eval_set=(X_val, y_val),
    plot=True                    #   
)

# 1.    
print("\n=== Validation Metrics ===")
y_val_pred = clf.predict(X_val)
print(f"Accuracy: {accuracy_score(y_val, y_val_pred):.4f}")
print("Classification Report:")
print(classification_report(y_val, y_val_pred))

#    
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_val, y_val_pred),
            annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Validation Confusion Matrix')
plt.show()

# 2.     
print("\n=== Test Metrics ===")
y_test_pred = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, y_test_pred))

#    
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_test_pred),
            annot=True, fmt='d', cmap='Greens',
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title('Test Confusion Matrix')
plt.show()

# 3.   ( )
if len(X_train.columns) <= 30:  #       
    plt.figure(figsize=(10, 6))
    sns.barplot(x=clf.feature_importances_, y=X_train.columns)
    plt.title('Feature Importance')
    plt.show()
```
